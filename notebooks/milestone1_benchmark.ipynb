```python
import pandas as pd

comparison_df = pd.DataFrame([
    {
        "Scenario": "Normal Detection",
        "Baseline Ratio (x)": result_normal["baseline_ratio"],
        "Selective Ratio (x)": result_normal["selective_ratio"],
        "Selective PSNR (dB)": result_normal["selective_psnr"],
        "Selective SSIM": result_normal["selective_ssim"],
        "Foreground Coverage (%)": result_normal["avg_fg_pct"],
        "Targets Detected": result_normal["has_targets"],
        "CRF Used": result_normal["crf_used"],
    },
    {
        "Scenario": "No Foreground Detected",
        "Baseline Ratio (x)": result_no_fg["baseline_ratio"],
        "Selective Ratio (x)": result_no_fg["selective_ratio"],
        "Selective PSNR (dB)": result_no_fg["selective_psnr"],
        "Selective SSIM": result_no_fg["selective_ssim"],
        "Foreground Coverage (%)": result_no_fg["avg_fg_pct"],
        "Targets Detected": result_no_fg["has_targets"],
        "CRF Used": result_no_fg["crf_used"],
    }
])

comparison_df
```

```python
def check_targets(result):
    return {
        "Compression >= 3x": result["selective_ratio"] >= 3.0,
        "PSNR >= 30 dB": result["selective_psnr"] >= 30.0,
        "SSIM >= 0.85": result["selective_ssim"] >= 0.85,
    }

print("=== Normal Detection Target Check ===")
print(check_targets(result_normal))

print("\n=== No Foreground Detected Target Check ===")
print(check_targets(result_no_fg))
```

# Milestone 1 Benchmark Summary

This notebook evaluates the selective compression pipeline on a local sample clip.

## Observations
- In the normal detection scenario, the pipeline preserved high visual quality but did not outperform the baseline in compression ratio.
- In the no-foreground scenario, the pipeline achieved a much higher compression ratio, but with lower image quality.
- These results show that the current implementation behaves adaptively based on scene content, but the current clip-level decision logic limits performance in mixed scenes.

## Conclusion
The benchmark confirms that PSNR, SSIM, and compression ratio can be measured end-to-end through the current pipeline. The results also highlight the tradeoff between image quality and compression efficiency.
